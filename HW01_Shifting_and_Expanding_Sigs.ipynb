**Question 1**:
You are given the graph of a signal in the time domain:


Untitled.jpg


Draw the transformations of the signal for each of the following:

- f(t - 1)
- f(2 - t)
- f(2t + 1)
- f(4 - t/2)



```python
import numpy as np
from matplotlib import pyplot as plt
```

let's define some needed functions:

```python
step = lambda t : t>0
ramp = lambda t : np.where(t>=0,t,0)
```

```python
N = 10000
lower_lim = -5
upper_lim = 5

t = np.linspace(lower_lim,upper_lim,N)
f = lambda t: ramp(t+2) - step(t+2) - ramp(t+1) + step(t+1) + step(t) - ramp(t-1) - step(t-1) + ramp(t-2)
```

**Original Signal f(t)**

```python
# 1. f(t)
plt.figure(figsize=(10,4))
plt.plot(t,f(t))
plt.title("Original Sig")
plt.xlabel("t [sec]")
plt.grid()
```

**f(t-1)**

```python
#2. f(t-1)

plt.figure(figsize=(10,4))
plt.plot(t,f(t-1))
plt.title("f(t-1)")
plt.xlabel("t [sec]")
plt.grid()
```

**f(2-t)**
1. first step: invert axis f(-t)
2. second step: move two units to the right (on inverted axis) -> f(-(t-2)) = f(2-t)

```python
#3. f(-t)
plt.figure(figsize=(10,4))
plt.plot(t,f(-t))
plt.title("f(-t)")
plt.xlabel("t [sec]")
plt.grid()
```

```python
#3. f(2-t)
plt.figure(figsize=(10,4))
plt.plot(t,f(2-t))
plt.title("f(2-t)")
plt.xlabel("t [sec]")
plt.grid()
```
